## Removing Outliers
### i) Removing floor_area_sqm outliers

```python
df15['floor_area_sqm'].describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(df15.floor_area_sqm, bins=20, rwidth=0.8)
plt.xlabel('Floor_area_sqm')
plt.ylabel('Frequency')

plt.show()
```

```python
# Calculate the IQR
Q1 = df15['floor_area_sqm'].quantile(0.25)
Q3 = df15['floor_area_sqm'].quantile(0.75)
IQR = Q3 - Q1


lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

print(lower_limit, upper_limit)
```

```python
floor_area_outliers = df15[(df15['floor_area_sqm']<lower_limit)|(df15['floor_area_sqm']>upper_limit)]
floor_area_outliers
```

```python
df16 = df15[(df15['floor_area_sqm']>=lower_limit)&(df15['floor_area_sqm']<=upper_limit)]
df16
```

### ii) Removing floor outlier

```python
df16['floor'].describe()
```

```python
df16['floor'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.hist(df16.floor, bins=20, rwidth=0.8)
plt.xlabel('Floor')
plt.ylabel('Frequency')

plt.show()
```

```python
threshold_floor = 30
floor_outlier = df16[df16['floor'] > threshold_floor]
floor_outlier
```

```python
df17 = df16[df16['floor'] <= threshold_floor]
df17
```

### iii) Removing resale price outliers

```python
df17['resale_price'].value_counts()
```

```python
df17['resale_price'].describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(df17.resale_price, bins=20, rwidth=0.8)

plt.xlabel('Resale Price')
plt.ylabel('Frequency')

plt.show()
```

```python
# Using z score method

from scipy import stats

# Calculate Z-scores
z_scores_price = stats.zscore(df17['resale_price'])

# Set a threshold (e.g., 3 standard deviations)
threshold_price = 3

# Identify outliers
resale_price_outliers = df17['resale_price'][abs(z_scores_price) > threshold_price]
resale_price_outliers
```

```python
df18 = df17[abs(z_scores_price) <= threshold_price]
df18
```

## Handling Address column

```python
df19 = df18.copy()
df19["address"].value_counts()
```

```python
df19["address"].unique()
```

```python
len(df19["address"].unique())
```

```python
top_20_addresses = df19['address'].value_counts().nlargest(20).index
top_20_addresses
```

```python
df19['address_encoded'] = df19['address'].where(df19['address'].isin(top_20_addresses), 'Other')
df19['address_encoded']
```

```python
df19
```

```python
df19["address_encoded"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode the 'address_encoded' column
le = LabelEncoder()
df19['address_encoded_numeric'] = le.fit_transform(df19['address_encoded'])
df19['address_encoded_numeric'].unique()
```

```python
# Create a DataFrame to map original addresses to their numeric labels
address_mapping = pd.DataFrame({
    'address_encoded': df19['address_encoded'].unique(),
    'address_encoded_numeric': df19['address_encoded_numeric'].unique()
})

# Display the mapping DataFrame
address_mapping
```

```python
df19
```

```python
df20 = df19.drop(["address", "address_encoded"], axis="columns")
df20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df20.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
